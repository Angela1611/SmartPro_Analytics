# Limpieza de Dataset sobre lesiones

```python
import pandas as pd
import numpy as np
import os
```

```python
ruta = '../new_datasets/Injuries.csv'
injuries = pd.read_csv(ruta, sep=',')
```

```python
#Eliminar datos previos a 2013
injuries = injuries[injuries['SEASON'] >= '12-13']
```

```python
injuries.shape
```

```python
Nan = injuries.isnull().sum()
Nan = Nan[Nan > 0]
injuries.shape[0] - Nan
```

```python
#Descripcion
def descripcion_datos(data):
    for i in data.columns:
        if data[i].dtype=='float64' or data[i].dtype=='int64':
            print(i)
            print('Mediana: ',data[i].median())
            print('Varianza: ',data[i].var())
            print('Rango: ',data[i].max()-data[i].min())
            print('Moda:')
            print(data[i].mode())
            print('------------------------------------')

    print('-------------------Datos nulos:-------------------')
    print(data.isnull().sum())
    print('-----------------------------------------')
    print('Dimensiones del dataframe:')
    print(np.shape(data))
    print('-----------------------------------------')

descripcion_datos(injuries)
```
