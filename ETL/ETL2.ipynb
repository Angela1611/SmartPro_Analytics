```python
def descripcion_datos(data):
    for i in data.columns:
        if data[i].dtype=='float64' or data[i].dtype=='int64':
            print(i)
            print('Mediana: ',data[i].median())
            print('Varianza: ',data[i].var())
            print('Rango: ',data[i].max()-data[i].min())
            print('Moda:')
            print(data[i].mode())
            print('------------------------------------')

    print('-------------------Datos nulos:-------------------')
    print(data.isnull().sum())
    print('-----------------------------------------')
    print('Dimensiones del dataframe:')
    print(np.shape(data))
    print('-----------------------------------------')


```

Descripción de las Columnas de la Tabla game
season_id (BIGINT): Identificador único para la temporada en la que se jugó el partido. Generalmente, esto puede ser un año o un código específico de temporada.

team_id_home (BIGINT): Identificador único del equipo local. Este ID se refiere al equipo que juega en su cancha.

team_abbreviation_home (VARCHAR(10)): Abreviatura del equipo local. Por ejemplo, "LAL" para Los Angeles Lakers.

team_name_home (VARCHAR(100)): Nombre completo del equipo local. Por ejemplo, "Los Angeles Lakers".

game_id (BIGINT): Identificador único para el partido. Este ID es específico para cada juego.

game_date (TIMESTAMP): Fecha y hora del partido. Este campo almacena cuándo se jugó el partido.

matchup_home (VARCHAR(100)): Descripción del enfrentamiento del equipo local. Generalmente en el formato "Equipo Local vs Equipo Visitante".

wl_home (VARCHAR(2)): Resultado del partido para el equipo local. "W" indica una victoria, "L" indica una derrota.

min (BIGINT): Minutos jugados en el partido. Suele ser 48 minutos para un partido estándar de la NBA.

fgm_home (FLOAT): Tiros de campo realizados por el equipo local. Es el número total de canastas realizadas.

fga_home (FLOAT): Tiros de campo intentados por el equipo local. Es el número total de canastas intentadas.

fg_pct_home (FLOAT): Porcentaje de tiros de campo realizados por el equipo local. Calculado como (fgm_home / fga_home) * 100.

fg3m_home (FLOAT): Triples realizados por el equipo local. Número de canastas de tres puntos realizadas.

fg3a_home (FLOAT): Triples intentados por el equipo local. Número de canastas de tres puntos intentadas.

fg3_pct_home (FLOAT): Porcentaje de triples realizados por el equipo local. Calculado como (fg3m_home / fg3a_home) * 100.

ftm_home (FLOAT): Tiros libres realizados por el equipo local. Número de tiros libres convertidos.

fta_home (FLOAT): Tiros libres intentados por el equipo local. Número de tiros libres intentados.

ft_pct_home (FLOAT): Porcentaje de tiros libres realizados por el equipo local. Calculado como (ftm_home / fta_home) * 100.

oreb_home (FLOAT): Rebotes ofensivos del equipo local. Número de rebotes obtenidos en el lado ofensivo de la cancha.

dreb_home (FLOAT): Rebotes defensivos del equipo local. Número de rebotes obtenidos en el lado defensivo de la cancha.

reb_home (FLOAT): Rebotes totales del equipo local. Suma de rebotes ofensivos y defensivos.

ast_home (FLOAT): Asistencias del equipo local. Número de pases que resultan en una canasta.

stl_home (FLOAT): Robos del equipo local. Número de veces que el equipo local le quita el balón al oponente.

blk_home (FLOAT): Bloqueos del equipo local. Número de tiros del oponente que son bloqueados.

tov_home (FLOAT): Pérdidas de balón del equipo local. Número de veces que el equipo local pierde la posesión del balón.

pf_home (FLOAT): Faltas personales del equipo local. Número de faltas cometidas por los jugadores del equipo local.

pts_home (FLOAT): Puntos anotados por el equipo local. Número total de puntos anotados por el equipo local.

plus_minus_home (BIGINT): Estadística de plus-minus para el equipo local. Representa la diferencia de puntos cuando el jugador o el equipo está en la cancha.

video_available_home (BIGINT): Indicador de disponibilidad de video para el equipo local. Puede ser 1 si el video está disponible y 0 si no lo está.

team_id_away (BIGINT): Identificador único del equipo visitante.

team_abbreviation_away (VARCHAR(10)): Abreviatura del equipo visitante.

team_name_away (VARCHAR(100)): Nombre completo del equipo visitante.

matchup_away (VARCHAR(100)): Descripción del enfrentamiento del equipo visitante.

wl_away (VARCHAR(2)): Resultado del partido para el equipo visitante. "W" para victoria, "L" para derrota.

fgm_away (FLOAT): Tiros de campo realizados por el equipo visitante.

fga_away (FLOAT): Tiros de campo intentados por el equipo visitante.

fg_pct_away (FLOAT): Porcentaje de tiros de campo realizados por el equipo visitante.

fg3m_away (FLOAT): Triples realizados por el equipo visitante.

fg3a_away (FLOAT): Triples intentados por el equipo visitante.

fg3_pct_away (FLOAT): Porcentaje de triples realizados por el equipo visitante.

ftm_away (FLOAT): Tiros libres realizados por el equipo visitante.

fta_away (FLOAT): Tiros libres intentados por el equipo visitante.

ft_pct_away (FLOAT): Porcentaje de tiros libres realizados por el equipo visitante.

oreb_away (FLOAT): Rebotes ofensivos del equipo visitante.

dreb_away (FLOAT): Rebotes defensivos del equipo visitante.

reb_away (FLOAT): Rebotes totales del equipo visitante.

ast_away (FLOAT): Asistencias del equipo visitante.

stl_away (FLOAT): Robos del equipo visitante.

blk_away (FLOAT): Bloqueos del equipo visitante.

tov_away (FLOAT): Pérdidas de balón del equipo visitante.

pf_away (FLOAT): Faltas personales del equipo visitante.

pts_away (FLOAT): Puntos anotados por el equipo visitante.

plus_minus_away (BIGINT): Estadística de plus-minus para el equipo visitante.

video_available_away (BIGINT): Indicador de disponibilidad de video para el equipo visitante.

season_type (VARCHAR(20)): Tipo de temporada, como "Regular", "Playoffs", etc.
